```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
df = pd.read_csv(url, sep=';')

print("Dataset berhasil dimuat. Berikut 5 baris pertama:")
display(df.head())

print("\nInformasi dasar dataset:")
df.info()

print("\nStatistik deskriptif dataset:")
display(df.describe())
```

### 2. Memeriksa Nilai yang Hilang

Mari kita periksa apakah ada nilai yang hilang dalam dataset.

```python
print("Jumlah nilai yang hilang per kolom:")
display(df.isnull().sum())
```

### 3. Analisis Variabel Target: `quality`

Mari kita lihat distribusi nilai kolom `quality`.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.countplot(x='quality', data=df)
plt.title('Distribusi Kualitas Anggur')
plt.xlabel('Kualitas')
plt.ylabel('Jumlah')
plt.show()

print("Jumlah kemunculan setiap nilai kualitas:")
display(df['quality'].value_counts().sort_index())
```

### 4. Distribusi Variabel Numerik

Selanjutnya, mari kita visualisasikan distribusi dari beberapa variabel numerik penting menggunakan histogram.

```python
df.hist(bins=20, figsize=(15, 10))
plt.suptitle('Distribusi Variabel Numerik')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
features = ['alcohol', 'pH', 'residual sugar', 'citric acid']

for feature in features:
    median_value = df[feature].median()

    df[feature + '_category'] = np.where(
        df[feature] >= median_value,
        'high',
        'low'
    )
```

```python
for feature in features:
    result = df.groupby(feature + '_category')['quality'].mean()

    print(feature)
    print(result)
    print()
```
