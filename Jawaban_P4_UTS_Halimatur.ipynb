```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load Dataset

```python
# Membaca dataset
df = pd.read_csv('/content/winequality-red.csv')

# Menampilkan 5 data pertama
print(df.head())
```

Menampilkan Informasi Dataset

```python
# Informasi dataset
print(df.info())

# Ukuran dataset
print("Jumlah baris dan kolom:", df.shape)
```

Mengecek Missing Value

```python
# Mengecek missing value
print(df.isnull().sum())
```

Statistik Deskriptif

```python
# Statistik deskriptif
print(df.describe())
```

Membuat Kolom Kategori

```python
# Membuat kategori high dan low berdasarkan median

df['alcohol_category'] = np.where(
    df['alcohol'] >= df['alcohol'].median(),
    'high',
    'low'
)

df['pH_category'] = np.where(
    df['pH'] >= df['pH'].median(),
    'high',
    'low'
)

df['residual_sugar_category'] = np.where(
    df['residual sugar'] >= df['residual sugar'].median(),
    'high',
    'low'
)

df['citric_acid_category'] = np.where(
    df['citric acid'] >= df['citric acid'].median(),
    'high',
    'low'
)
```

Analisis Rata-rata Quality


Alcohol

```python
# Analisis alcohol
alcohol_analysis = df.groupby('alcohol_category')['quality'].mean()

print("alcohol")
print(alcohol_analysis)
```

pH

```python
# Analisis pH
ph_analysis = df.groupby('pH_category')['quality'].mean()

print("pH")
print(ph_analysis)
```

Residual Sugar

```python
# Analisis residual sugar
sugar_analysis = df.groupby('residual_sugar_category')['quality'].mean()

print("residual_sugar")
print(sugar_analysis)
```

Citric Acid

```python
# Analisis citric acid
citric_analysis = df.groupby('citric_acid_category')['quality'].mean()

print("citric_acid")
print(citric_analysis)
```

Menampilkan Semua Hasil Sekaligus

```python
print("\n===== HASIL ANALISIS =====\n")

print("alcohol")
print(df.groupby('alcohol_category')['quality'].mean())

print("\npH")
print(df.groupby('pH_category')['quality'].mean())

print("\nresidual_sugar")
print(df.groupby('residual_sugar_category')['quality'].mean())

print("\ncitric_acid")
print(df.groupby('citric_acid_category')['quality'].mean())
```

B. EXPLORATORY DATA ANALYSIS (EDA)

1. Distribusi Quality

```python
plt.figure(figsize=(8,5))
sns.countplot(x='quality', data=df)

plt.title('Distribusi Quality Wine')
plt.xlabel('Quality')
plt.ylabel('Jumlah Data')

plt.show()
```

2. Histogram Semua Fitur

```python
# Histogram semua fitur
df.hist(figsize=(15,12))

plt.tight_layout()
plt.show()
```

3. Korelasi Antar Fitur

```python
# Korelasi
corr = df.corr(numeric_only=True)

plt.figure(figsize=(12,8))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```

4. Visualisasi Alcohol vs Quality

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='quality',
    y='alcohol',
    data=df
)

plt.title('Alcohol vs Quality')

plt.show()
```

5. Visualisasi Citric Acid vs Quality

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='quality',
    y='citric acid',
    data=df
)

plt.title('Citric Acid vs Quality')

plt.show()
```

6. Scatter Plot Alcohol dan Quality

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='alcohol',
    y='quality',
    data=df
)

plt.title('Alcohol vs Quality')

plt.show()
```

7. Scatter Plot pH dan Quality

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='pH',
    y='quality',
    data=df
)

plt.title('pH vs Quality')

plt.show()
```

8. Pairplot

```python
sns.pairplot(
    df[['alcohol', 'pH', 'citric acid', 'residual sugar', 'quality']]
)

plt.show()
```
